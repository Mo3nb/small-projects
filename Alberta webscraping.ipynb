```python
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
import pandas as pd
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver = webdriver.Safari()
url="https://www.volunteerconnector.org/?pc=H3L&md=5000&se=&so=Newest"
href=[]
nextpage=[]
Websites=[]
while True:
    try:
        driver.get(url)
        html_content = requests.get(url).text
        soup = BeautifulSoup(html_content, "lxml")
        table = soup.find('div', attrs={'class':"column opportunity-list"})
        webs=table.findAll('a')
        prefix_url = 'https://www.volunteerconnector.org'
        nextpage=[]
        for link in webs:
            if len(link) >1:
                href.append(link.get("href"))
                #print("href: {}".format(link.get("href")))
                
        driver.find_element_by_link_text('Next').click()
        WebDriverWait(driver, 15).until(EC.url_changes(url))
        url=driver.current_url
    except:  
        break
        
        
for i in range(len(href)):
   # print("Position: {}".format(position[i]))
    Websites.append(prefix_url+href[i])
    print("Website: {}".format(Websites[i]))
driver.quite()
        
```

```python
Websites[0]
```

```python
title=[]
COVID=[]
address=[]
duration=[]
date=[]
description=[]
organization=[]
category=[]
requirement=[]
for link in Websites:

    html_content2 = requests.get(link).text
    soup2 = BeautifulSoup(html_content2, "lxml")
    blocks = soup2.find('section', attrs={'class':"column right"})
    if blocks.find('div', attrs={'class':"covid-19 button is-small is-primary"}) is not None:
        COVID.append("yes")
    else:
        COVID.append("No")

    title.append(blocks.find('h1').text)
    address.append(blocks.find('div' , class_="location").text.strip())
    if blocks.find('div' , class_="duration") is not None:
        duration.append(blocks.find('div' , class_="duration").text)
    else:
        duration.append('None')
    date.append(blocks.find('div' , class_="dates").text)
    descr=blocks.find('div' , class_="description").text
    if blocks.find('div', attrs={'class':"additional-details"}) is not None:
        addi=blocks.find('div', attrs={'class':"additional-details"})
        additional=addi.find('div' , class_="description").text      
    if blocks.find('div', attrs={'class':"why-volunteer"})is not None:
        addit=blocks.find('div', attrs={'class':"why-volunteer"})
        additional2=addit.find('div' , class_="description").text
    description.append(descr+additional+additional2)
    left=soup2.find('section', attrs={'class':"column is-3 left"})
    organization.append(left.find('a' , class_="org-link").text)
    if left.find('div' , class_="activities is-hidden-touch") is not None:
        category.append(left.find('div' , class_="activities is-hidden-touch").text[21:]) 
    else:
        category.append('None')
    if left.find('div' , class_="requirements is-hidden-touch") is not None:
        requirement.append(left.find('div' , class_="requirements is-hidden-touch").text[16:])
    else:
        requirement.append("None")
```

### Output CSV file

```python
len(COVID)
```

```python
df = pd.DataFrame({'Position':title,
                   'Organization':organization,
                   "Websites": Websites,
                   "Description": description,
                   "Qualification": requirement,
                   'Date_Posted':"",
                   "Deadline": "",
                   'Location':address,
                   "Category": category,
                   "Goal":"",
                   "Time": date,
                   "Training" :"",
                   "COVID":COVID,
                   'workhour':duration,
                  }) 
```

```python
df.to_csv('alberta_Volunteer_all.csv', index=False, encoding='utf-8_sig')
```
